## Project description
The telecom operator "TeleDom" wants to combat customer churn. To achieve this, its employees will start offering promo codes and special conditions to everyone planning to discontinue telecom services. To identify such users in advance, TeleDom needs a model that will predict whether a subscriber will terminate their contract. The operator's team has collected personal data about some customers, information about their tariffs and services.

Our task is **to train a model on this data to forecast customer churn**.

### Service description

The operator provides two main types of services: 
1) Landline phone service. The phone can be connected to multiple lines simultaneously.  
2) Internet. Connection is available in two types: through a telephone line (DSL) or fiber optic cable. 

Subscribers also have access to a range of services:
- Internet security: antivirus (Device Protection) and blocking of dangerous websites (Online Security);
- Dedicated technical support line (Tech Support);
- Cloud file storage for data backup (Online Backup);
- Streaming TV and movie catalog (Streaming Movies).

### Brief table descriptions
The data is stored in a PostgreSQL database. It consists of several tables:
1) contract — contract information;  
2) personal — customers' personal data;  
3) internet — internet service information;  
4) phone — telephony service information.

```python
# Merging tables
df = contract.merge(personal, on ='customer_id', how='inner')
df = df.merge(internet, on ='customer_id', how = 'outer')
df = df.merge(phone, on ='customer_id', how = 'outer')
df.name = 'df'
df_info(df, 'df')
```

#### Eliminating missing values

```python
# Transform the end_date column into our target variable
def create_target(row):
    if row['end_date'] is None:
        return 0
    return 1

df['target'] = df.apply(create_target, axis=1)
```

```python
# Change the values in the senior_citizen column from int to object
df['senior_citizen'] = df['senior_citizen'].astype(object)
```

A hypothesis was put forward that the missing values in the total_charges column are related to the fact that the person became a company user on February 1, 2022 (on the day when the data were collected and exported).

```python
# Let's examine the missing values in the total_charges column
df.loc[df['total_charges'].isna()]
```

Our hypothesis was confirmed. The missing values are related to people who signed a contract with "TeleDom" on the last day of data collection, so it was decided to multiply the monthly charge by the "payment type" and obtain the total payment.

```python
# Fill all missing values in total_charges
def fillna_train(row):
    if pd.isnull(row['total_charges']):
        if row['type'] == 'Two year':
            row['total_charges'] = row['monthly_charges'] * 24
        elif row['type'] == 'TOne year':
            row['total_charges'] = row['monthly_charges'] * 12
        else:
            row['total_charges'] = row['monthly_charges']
    else:
        row['total_charges'] = row['total_charges']
    return row

df = df.apply(fillna_train, axis=1)
```

```python
# Checking the correctness of the function's operation
df.head()
```

```python
# Fill NaN values in the remaining columns
phone = ['multiple_lines']
internet = ['internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_TV', 'streaming_movies']

df[phone] = df[phone].fillna('no phone')
df[internet] = df[internet].fillna('no internet')
```

```python
# Convert begin_date and end_date to date format
df['begin_date'] = pd.to_datetime(df['begin_date'])
df['end_date'] = pd.to_datetime(df['end_date'])
```

```python
def days_until_payment(row):
    type = row['type']
    duration_days = row['days']
    # Get the remaining number of days until the payment deadline depending on its type
    # Use division with remainder
    if type == 'Month-to-month':
        return 30- (duration_days % 30)
    elif type == 'One year':
        return 365 - (duration_days % 365)
    else:
        return 730 - (duration_days % 730)
```

```python
df['days_until_payment'] = df.apply(days_until_payment, axis=1)
```

```python
plt.figure(figsize = (10, 5))
sns.violinplot(x="target", y="days_until_payment", hue="gender",
                   data=df, palette="coolwarm", dodge=True)
plt.title('Graph of the relationship between the number of days until payment and the chance that a client will refuse the service', fontsize = 15, pad = 15)
plt.show()
print()
print('-' * 50, 'Correlation:', np.round(df['target'].corr(df['days_until_payment']), 2), '-' * 50)
```

Now we can clearly see that the chance that a client will refuse the tariff significantly increases when the tariff payment day approaches.

####  Model 6. Neural Network

```python
# Convert samples to tensors
X_train = torch.FloatTensor(np.array(features_train_log))
X_test = torch.FloatTensor(np.array(features_test_log))
y_train = torch.LongTensor(np.array(target_train))
y_test = torch.LongTensor(np.array(target_test))
```

```python
# Add Batch Normalization and Dropout regularization 
class Net_With_DP(nn.Module):
    def __init__(self, n_in_neurons, n_hidden_neurons_1, n_hidden_neurons_2, n_hidden_neurons_3, n_out_neurons, p):
            super(Net_With_DP, self).__init__()
            
            self.fc1 = nn.Linear(n_in_neurons, n_hidden_neurons_1)
            self.bn1 = nn.Dropout(p = p)
            self.act1 = nn.Tanh()
            self.fc2 = nn.Linear(n_hidden_neurons_1, n_hidden_neurons_2)
            self.dp2 = nn.Dropout(p = p)
            self.act2 = nn.ReLU()
            self.fc3 = nn.Linear(n_hidden_neurons_2, n_hidden_neurons_3)
            self.act3 = nn.ELU()
            self.fc4 = nn.Linear(n_hidden_neurons_3, n_out_neurons)
            
    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.act1(x)
        x = self.fc2(x)
        x = self.dp2(x)
        x = self.act2(x)
        x = self.fc3(x)
        x = self.act3(x)
        x = self.fc4(x)
        
        return x
```

```python
X_train.shape[1]
```

```python
n_in_neurons = X_train.shape[1]
n_hidden_neurons_1 = 17
n_hidden_neurons_2 = 11
n_hidden_neurons_3 = 5
n_out_neurons = 2
```

```python
class Early_stopping_counter():
    def __init__(self, patience = 20, min_delta = 0):
        self.patience = patience
        self.min_delta = min_delta
        self.counter = 0
        self.best_loss = None
        self.early_stop = False

    def __call__(self, val_loss, current_metric):
        if self.best_loss == None:
            self.best_loss = val_loss
        elif self.best_loss - val_loss > self.min_delta:
            self.best_loss = val_loss
            self.counter = 0
        elif self.best_loss - val_loss < self.min_delta:
            self.counter += 1
            print(f"Early stopping counter {self.counter} of {self.patience}, Current ROC_AUC: {current_metric}")
            if self.counter >= self.patience:
                print('FINAL: Early stopping')
                self.early_stop = True
```

```python
# Create dataset
train_dataset = torch.utils.data.TensorDataset(X_train, y_train)
test_dataset = torch.utils.data.TensorDataset(X_test, y_test)
```

```python
def loader(batch):
    train_loader = DataLoader(train_dataset, batch_size = batch, shuffle=True, num_workers=0)
    test_loader = DataLoader(test_dataset, batch_size = batch, num_workers=0)
    return train_loader, test_loader
```

```python
def train_with_early_stopping(train_loader, test_loader, num_epochs=1000001, verbose=1):
    
    for epoch in range(num_epochs):
        net.train()
        for batch in train_loader:
            Xt, yt = batch
            optimizer.zero_grad()
            preds = net.forward(Xt) 
            loss_value = loss(preds, yt)
            loss_value.backward()
            optimizer.step()

        if epoch % verbose == 0:
            predicted_value = []
            true_values = []  
            with torch.no_grad():
                net.eval()
                for batch in test_loader:
                    Xtest, ytest = batch
                    test_preds = net.forward(Xtest) 
                    predicted_value.append(test_preds)
                    true_values.append(ytest)
            
            # Process predictions for ROC-AUC
            predicted_value = torch.cat(predicted_value).detach()
            true_values = torch.cat(true_values).detach()
            
            # For binary classification with CrossEntropyLoss, get probabilities
            probabilities = torch.softmax(predicted_value, dim=1)[:, 1].numpy()
            true_values_np = true_values.numpy()
            
            AUC = roc_auc_score(true_values_np, probabilities)
            
            early_stopping(AUC, AUC)

            if early_stopping.counter == 0:
                best_auc = AUC
                best_predicted = predicted_value
                print(f"epoch:{epoch}, AUC test: {AUC}")

            if early_stopping.early_stop:
                print()
                print('Early Stopping!!!')
                print(f'The best AUC test {best_auc}')
                
                break

    return best_predicted, best_auc
```

```python
def net_init(n_in_neurons, n_hidden_neurons_1, n_hidden_neurons_2, 
                                      n_hidden_neurons_3, n_out_neurons, p = 0):
    
    torch.manual_seed(RS)
    
    net = Net_With_DP(n_in_neurons, n_hidden_neurons_1, n_hidden_neurons_2, 
                                      n_hidden_neurons_3, n_out_neurons, p = p_dropout)
    
    optimizer = torch.optim.Adam(net.parameters(), lr=1e-3)
    
    early_stopping = Early_stopping_counter(patience = 10, min_delta = 0)
    
    loss = nn.CrossEntropyLoss()
    
    return net, optimizer, early_stopping, loss
```

```python
dfh = pd.DataFrame(columns=['batch_size', 'p_dropout', 'best_auc', 'best_predicted'])

# Hyperparameters for search
batch_list = [32, 48]
p_dropout_list = [0.3, 0.5, 0.6, 0.7]

k_folds = 3
kfold = KFold(n_splits=k_folds, shuffle=True)

for fold, (train_ids, test_ids) in enumerate(kfold.split(train_dataset)):

    # Print
    print(f'FOLD {fold}')
    print('--------------------------------')

    for batch in batch_list:
        for p_dropout in p_dropout_list:

            print(f'--- batch_size = {batch}, p_dropout = {p_dropout} ---')

            # Initialize the next model
            net, optimizer, early_stopping, loss = net_init(n_in_neurons, n_hidden_neurons_1, n_hidden_neurons_2, 
                                                                      n_hidden_neurons_3, n_out_neurons, p = p_dropout)    
            # Dataloader for training and test parts of the dataset
            train_loader, test_loader = loader(batch)

            # Model training
            best_predicted, best_auc = train_with_early_stopping(train_loader, test_loader, 
                                                                          num_epochs=1000001, verbose=1)
    
            # Add the results
            dfh.loc[len(dfh)] = [batch, p_dropout, best_auc.item(), best_predicted]
```
